가입했는지 갱신했는지 알기 쉽게 숫자로 된 데이터를 한글로 바꿉니다.

```python
df2 = df.copy()
df2.columns = ['가입해지여부', '가입기간', '가입갱신여부', '데이터요금제사용여부', '데이터사용량(GB)', '고객센터전화량', '월평균전화량(분)', '월평균전화수(횟수)', '월평균요금', '초과요금', '로밍시간(분)']
df2['가입해지여부'] = df2['가입해지여부'].replace(0, '가입')
df2['가입해지여부'] = df2['가입해지여부'].replace(1, '해지')
df2['가입갱신여부'] = df2['가입갱신여부'].replace(1, '갱신')
df2['가입갱신여부'] = df2['가입갱신여부'].replace(0, '비갱신')
df2['데이터요금제사용여부'] = df2['데이터요금제사용여부'].replace(1, '사용')
df2['데이터요금제사용여부'] = df2['데이터요금제사용여부'].replace(0, '사용안함')
```

가입, 갱신, 데이터사용 이 세 가지를 모두 사용하는 데이터의 통계를 구합니다.

```python
#drop은 자료를 없애고 inplace=True는 원본 데이터에 변경사항을 반영하게 합니다.
df3 = df2.copy()
indexNames = df3[ df3['가입해지여부'] == '해지' ].index
df3.drop(indexNames , inplace=True)
indexNames2 = df3[ df3['가입갱신여부'] == '비갱신' ].index
df3.drop(indexNames2 , inplace=True)
indexNames3 = df3[ df3['데이터요금제사용여부'] == '사용안함' ].index
df3.drop(indexNames3 , inplace=True)
df3
```

```python
#원하는 컬럼만 모아 분석합니다.
df_cal = df3[['가입기간', '데이터사용량(GB)', '고객센터전화량', '월평균전화량(분)', '월평균전화수(횟수)', '월평균요금', '초과요금', '로밍시간(분)']]
df_cal.describe().T
```

최대 월평균전화수는 163회 등 다양하게 데이터를 분석할 수 있습니다.

시각화를 합니다.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(font="Malgun Gothic")
```

자료가 방대한 관계로 데이터사용량이 4.3 이상인 자료를 위주로 시각화합니다.

```python
df4 = df2.copy()
indexNames5 = df4[df4['데이터사용량(GB)'] <= 4.3].index
df4.drop(indexNames5 , inplace=True)
df4
```

```python
#데이터사용량 기준으로 오름차순으로 정리합니다.
df4 = df4.sort_values(by='데이터사용량(GB)', ascending=False)
df4
```

```python
#ci=None은 신뢰구간을 제거합니다.
plt.figure(figsize=(17, 6))
sns.barplot(data=df4, x='데이터사용량(GB)', y='월평균요금', ci=None)
```

흥미로운 결과입니다. 데이터사용량과 월평균요금이 꼭 비례하지는 않다는 것을 알 수 있습니다.

```python
#hue를 이용해 월평균전화수도 한 번에 보게끔 할 수 있습니다.
plt.figure(figsize=(15, 6))
sns.pointplot(data=df4, x='데이터사용량(GB)', y='월평균전화량(분)', ci=None, hue='월평균전화수(횟수)')
#legend함수를 이용해 범례의 위치를 그래프를 안가리는 곳으로 옮길 수 있습니다.
plt.legend(loc='center right', bbox_to_anchor=(1.08, 0.5), ncol=1)
```

데이터사용량이 가장 큰 곳이 월평균전화량이 가장 많은 것도 아니고 월평균전화수가 많은 것도 아니라는 걸 알 수 있습니다.

즉, 데이터사용량과 월평균전화량, 월평균전화수는 비례하지 않다는 것입니다.

요금을 기준으로 데이터를 시각화합니다.

```python
cost = df2.sort_values(by='월평균요금', ascending=False)
cost = cost.head(100)
```

```python
count = df2.sort_values(by='월평균전화수(횟수)', ascending=False)
count = count.head(100)
```

```python
plt.figure(figsize=(15, 6))
sns.lineplot(data=count, x='데이터사용량(GB)', y='월평균전화수(횟수)', color='orange', label='월평균전화수', ci=None)
sns.lineplot(data=cost, x='데이터사용량(GB)', y='월평균요금', label='월평균요금', ci=None)
plt.ylabel('수치')
```

월평균요금이 가장 많은 데이터사용량은 4.43GB정도이고

월평균전화수가 많은 데이터사용량은 0과 3GB라는 것을 알 수 있습니다.

월평균요금 상위 5개를 시각화하여 분석합니다.

```python
cost2 = cost.head()
```

```python
plt.figure(figsize=(17, 7))
sns.barplot(data=cost2, x='월평균전화량(분)', y='월평균요금', ci=None, hue='초과요금')
plt.legend(loc='center right', bbox_to_anchor=(1.05, 0.65), ncol=1)
```

월평균요금이 가장 큰 월평균 사용량은 268.8분정도이며 초과요금은 12.33입니다.

```python
# 아래와 같이 하면 두 개의 그래프를 한 곳에 나타나게 할 수 있습니다.
# 한 곳에 두 개의 자료를 나타낼 때 라벨의 이름이 헷갈리므로 라벨의 이름을 지정합니다.
plt.figure(figsize=(15, 5))
sns.lineplot(data=cost, x='월평균요금', y='데이터사용량(GB)', label='데이터사용량')
sns.lineplot(data=cost, x='월평균요금', y='초과요금', color='orange', label='초과요금')
plt.ylabel('수치')
```
